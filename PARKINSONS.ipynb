UNIVARIATE ANALYSIS AND OUTLIERS DETECTION AND CAPPING

```python
import matplotlib.pyplot as plt
import seaborn as sns
selected_columns = df.select_dtypes(include=['int64','float64']).columns
plt.figure(figsize=(24, 4 * ((len(selected_columns) + 5) // 6)))
for i, col in enumerate(selected_columns):
  plt.subplot((len(selected_columns) + 5) // 6, 6, i + 1)
  sns.boxplot(y=df[col], color='lightgreen')
  plt.title(col)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

def cap_outliers_iqr(df, target_column='status'):
  columns = df.select_dtypes(include=['int64', 'float64']).columns
  columns = [col for col in columns if col != target_column]

  for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    if df[col].min() >= 0 and lower_bound < 0:
        lower_bound = 0
    outliers_before = len(df[(df[col] < lower_bound) | (df[col] > upper_bound)])

    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

    outliers_after = len(df[(df[col] < lower_bound) | (df[col] > upper_bound)])

    print(f"{col}: Capped below {lower_bound} and above {upper_bound} | "
          f"Original Min: {df[col].min()}, Max: {df[col].max()} | "
          f"Outliers before: {outliers_before}, Outliers after: {outliers_after}")
  return df
df = cap_outliers_iqr(df, target_column='status')
```

**MACHINE LEARNING MODELS**

ORIGINAL DATA (WITHOUT SAMPLING)

```python
X=df_copy.drop("status",axis=1)
y=df_copy["status"]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, confusion_matrix,f1_score, precision_score, recall_score, roc_auc_score)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import joblib
import warnings
warnings.filterwarnings("ignore")

#Drop missing values
X_clean = X.dropna()
y_clean = y.loc[X_clean.index]

#Train-test split
xtrain, xtest, ytrain, ytest = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42, stratify=y_clean)

#Feature Importance
rf_temp = RandomForestClassifier(random_state=42)
rf_temp.fit(xtrain, ytrain)

#Top 10 features
importances = rf_temp.feature_importances_
forest_importances = pd.Series(importances, index=X_clean.columns)
top10_features = forest_importances.sort_values(ascending=False).head(10).index.tolist()

print("\nTop 10 important features:")
print(top10_features)

#Plot feature importances
std = np.std([tree.feature_importances_ for tree in rf_temp.estimators_], axis=0)
fig, ax = plt.subplots(figsize=(10, 6))
forest_importances[top10_features].plot.bar(yerr=std[:10], ax=ax)
ax.set_title("Top 10 Feature Importances")
ax.set_ylabel("Importance")
plt.tight_layout()
plt.show()

#Scaling Only Top 10 Features
scaler = MinMaxScaler()
xtrain_top = pd.DataFrame(scaler.fit_transform(xtrain[top10_features]), columns=top10_features)
xtest_top = pd.DataFrame(scaler.transform(xtest[top10_features]), columns=top10_features)

#Train Final Model
rf_top = RandomForestClassifier(random_state=42)
rf_top.fit(xtrain_top, ytrain)
y_pred_top = rf_top.predict(xtest_top)

#Evaluation
print('\nTop 10 Feature Model Evaluation:')
print(classification_report(ytest, y_pred_top))
print('Accuracy:', accuracy_score(ytest, y_pred_top))
print('F1 Score:', f1_score(ytest, y_pred_top, average='weighted'))
print('Precision:', precision_score(ytest, y_pred_top, average='weighted'))
print('Recall:', recall_score(ytest, y_pred_top, average='weighted'))
y_proba = rf_top.predict_proba(xtest_top)[:, 1]
print('ROC-AUC:', roc_auc_score(ytest, y_proba))
print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred_top))
```

```python
#OTHER MODELS
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
models = {"Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
    "Naive Bayes": GaussianNB()}
for name, model in models.items():
  model.fit(xtrain_top, ytrain)
  y_pred = model.predict(xtest_top)
  try:
      if hasattr(model, "predict_proba"):
          y_scores = model.predict_proba(xtest_top)[:, 1]
      else:
          y_scores = model.decision_function(xtest_top)
  except:
      y_scores = y_pred

  print(f"\n {name} Evaluation")
  print(classification_report(ytest, y_pred))
  print('Accuracy     :', accuracy_score(ytest, y_pred))
  print('F1 Score     :', f1_score(ytest, y_pred, average='weighted'))
  print('Precision    :', precision_score(ytest, y_pred, average='weighted'))
  print('Recall       :', recall_score(ytest, y_pred, average='weighted'))
  try:
      print('ROC-AUC     :', roc_auc_score(ytest, y_scores))
  except:
      print('ROC-AUC      : Not available (check score format)')

  print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred))
```

CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from sklearn.metrics import make_scorer, f1_score
import numpy as np
top_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
scores = cross_val_score(top_model, xtrain_top, ytrain, cv=20, scoring='accuracy')
print("\nCross-validated Accuracy Scores:", scores)
print("Mean CV Accuracy:", np.mean(scores))
```

WITH UNDERSAMPLING

```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, confusion_matrix,f1_score, precision_score, recall_score, roc_auc_score)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import joblib
import warnings
warnings.filterwarnings("ignore")

X_clean = X.dropna()
y_clean = y.loc[X_clean.index]

rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X_clean, y_clean)
print('Resampled class distribution:', Counter(y_res))

xtrain, xtest, ytrain, ytest = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

rf_temp = RandomForestClassifier(random_state=42)
rf_temp.fit(xtrain, ytrain)

importances = rf_temp.feature_importances_
forest_importances = pd.Series(importances, index=X_res.columns)
top10_features = forest_importances.sort_values(ascending=False).head(10).index.tolist()

print("\n Top 10 Important Features:")
print(top10_features)

std = np.std([tree.feature_importances_ for tree in rf_temp.estimators_], axis=0)
fig, ax = plt.subplots(figsize=(10, 6))
forest_importances[top10_features].plot.bar(yerr=std[:10], ax=ax)
ax.set_title("Top 10 Feature Importances")
ax.set_ylabel("Importance")
plt.tight_layout()
plt.show()

scaler = MinMaxScaler()
xtrain_top = pd.DataFrame(scaler.fit_transform(xtrain[top10_features]), columns=top10_features)
xtest_top = pd.DataFrame(scaler.transform(xtest[top10_features]), columns=top10_features)

rf_top = RandomForestClassifier(random_state=42)
rf_top.fit(xtrain_top, ytrain)
y_pred_top = rf_top.predict(xtest_top)

print('\n Top 10 Feature Model Evaluation:')
print(classification_report(ytest, y_pred_top))
print('Accuracy:', accuracy_score(ytest, y_pred_top))
print('F1 Score:', f1_score(ytest, y_pred_top, average='weighted'))
print('Precision:', precision_score(ytest, y_pred_top, average='weighted'))
print('Recall:', recall_score(ytest, y_pred_top, average='weighted'))
y_proba = rf_top.predict_proba(xtest_top)[:, 1]
print('ROC-AUC:', roc_auc_score(ytest, y_proba))
print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred_top))
```

```python
#OTHER MODELS
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
models = {"Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
    "Naive Bayes": GaussianNB()}
for name, model in models.items():
  model.fit(xtrain_top, ytrain)
  y_pred = model.predict(xtest_top)
  try:
      if hasattr(model, "predict_proba"):
          y_scores = model.predict_proba(xtest_top)[:, 1]
      else:
          y_scores = model.decision_function(xtest_top)
  except:
      y_scores = y_pred

  print(f"\n{name} Evaluation ")
  print(classification_report(ytest, y_pred))
  print('Accuracy     :', accuracy_score(ytest, y_pred))
  print('F1 Score     :', f1_score(ytest, y_pred, average='weighted'))
  print('Precision    :', precision_score(ytest, y_pred, average='weighted'))
  print('Recall       :', recall_score(ytest, y_pred, average='weighted'))
  try:
      print('ROC-AUC      :', roc_auc_score(ytest, y_scores))
  except:
      print('ROC-AUC      : Not available (check score format)')

  print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred))
```
