UNIVARIATE ANALYSIS, OUTLIER DETECTION AND CAPPING

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
numeric_cols = df.select_dtypes(include='number').drop(columns=['Dataset']).columns
num_cols = 5
num_plots = len(numeric_cols)
num_rows = int(np.ceil(num_plots / num_cols))
plt.figure(figsize=(6 * num_cols, 4 * num_rows))
for i, col in enumerate(numeric_cols):
  plt.subplot(num_rows, num_cols, i + 1)
  sns.boxplot(y=df[col], color='lightgreen')
  plt.title(col)
  plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

def cap_outliers_iqr(df):
  numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
  for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    if lower_bound < 0:
        lower_bound = 0
    original_min = df[col].min()
    original_max = df[col].max()
    df.loc[df[col] < lower_bound, col] = lower_bound
    df.loc[df[col] > upper_bound, col] = upper_bound
    print(f"{col}: capped below {lower_bound:.2f} and above {upper_bound:.2f} | Original Min: {original_min:.2f}, Max: {original_max:.2f}")
  return df
df_capped= cap_outliers_iqr(df)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Gender'], df['Dataset'])
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
alpha = 0.05
if p_value <= alpha:
  print("\nThere is a significant relationship between gender and Dataset.")
else:
  print("\nThere is no significant relationship between gender and Dataset.")
```

ENCODING

```python
df_encoded['Gender'] = df_encoded['Gender'].map({'Male': 1, 'Female': 0})
```

```python
X=df_encoded.drop("Dataset",axis=1)
y=df_encoded["Dataset"]
y.value_counts()
```

ORIGINAL DATA (WITHOUT SAMPLING)

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, confusion_matrix,f1_score, precision_score, recall_score, roc_auc_score)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import joblib
import warnings
warnings.filterwarnings("ignore")

#Drop missing values
X_clean = X.dropna()
y_clean = y.loc[X_clean.index]

#Train-test split
xtrain, xtest, ytrain, ytest = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42, stratify=y_clean)

#Feature Importance
rf_temp = RandomForestClassifier(random_state=42)
rf_temp.fit(xtrain, ytrain)

#Top 6 features
importances = rf_temp.feature_importances_
forest_importances = pd.Series(importances, index=X_clean.columns)
top6_features = forest_importances.sort_values(ascending=False).head(6).index.tolist()

print("\nTop 6 important features:")
print(top6_features)

#Plot feature importances
std = np.std([tree.feature_importances_ for tree in rf_temp.estimators_], axis=0)
fig, ax = plt.subplots(figsize=(10, 6))
forest_importances[top6_features].plot.bar(yerr=std[:6], ax=ax)
ax.set_title("Top 6 Feature Importances")
ax.set_ylabel("Importance")
plt.tight_layout()
plt.show()

#Scaling Only Top 10 Features
scaler = MinMaxScaler()
xtrain_top = pd.DataFrame(scaler.fit_transform(xtrain[top6_features]), columns=top6_features)
xtest_top = pd.DataFrame(scaler.transform(xtest[top6_features]), columns=top6_features)

#Train Final Model
rf_top = RandomForestClassifier(random_state=42)
rf_top.fit(xtrain_top, ytrain)
y_pred_top = rf_top.predict(xtest_top)

#Evaluation
print('\nTop 6 Feature Model Evaluation:')
print(classification_report(ytest, y_pred_top))
print('Accuracy:', accuracy_score(ytest, y_pred_top))
print('F1 Score:', f1_score(ytest, y_pred_top, average='weighted'))
print('Precision:', precision_score(ytest, y_pred_top, average='weighted'))
print('Recall:', recall_score(ytest, y_pred_top, average='weighted'))
y_proba = rf_top.predict_proba(xtest_top)[:, 1]
print('ROC-AUC:', roc_auc_score(ytest, y_proba))
print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred_top))
```

```python
#OTHER MODELS
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
models = {"Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()}
for name, model in models.items():
  model.fit(xtrain_top, ytrain)
  y_pred = model.predict(xtest_top)
  try:
      if hasattr(model, "predict_proba"):
          y_scores = model.predict_proba(xtest_top)[:, 1]
      else:
          y_scores = model.decision_function(xtest_top)
  except:
      y_scores = y_pred

  print(f"\n {name} Evaluation")
  print(classification_report(ytest, y_pred))
  print('Accuracy     :', accuracy_score(ytest, y_pred))
  print('F1 Score     :', f1_score(ytest, y_pred, average='weighted'))
  print('Precision    :', precision_score(ytest, y_pred, average='weighted'))
  print('Recall       :', recall_score(ytest, y_pred, average='weighted'))
  try:
      print('ROC-AUC     :', roc_auc_score(ytest, y_scores))
  except:
      print('ROC-AUC      : Not available (check score format)')

  print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred))
```

WITH UNDERSAMPLING

```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, confusion_matrix,f1_score, precision_score, recall_score, roc_auc_score)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import joblib
import warnings
warnings.filterwarnings("ignore")

X_clean = X.dropna()
y_clean = y.loc[X_clean.index]

rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X_clean, y_clean)
print('Resampled class distribution:', Counter(y_res))

xtrain, xtest, ytrain, ytest = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

rf_temp = RandomForestClassifier(random_state=42)
rf_temp.fit(xtrain, ytrain)

importances = rf_temp.feature_importances_
forest_importances = pd.Series(importances, index=X_res.columns)
top6_features = forest_importances.sort_values(ascending=False).head(6).index.tolist()

print("\n Top 6 Important Features:")
print(top6_features)

std = np.std([tree.feature_importances_ for tree in rf_temp.estimators_], axis=0)
fig, ax = plt.subplots(figsize=(10, 6))
forest_importances[top6_features].plot.bar(yerr=std[:6], ax=ax)
ax.set_title("Top 6 Feature Importances")
ax.set_ylabel("Importance")
plt.tight_layout()
plt.show()

scaler = MinMaxScaler()
xtrain_top = pd.DataFrame(scaler.fit_transform(xtrain[top6_features]), columns=top6_features)
xtest_top = pd.DataFrame(scaler.transform(xtest[top6_features]), columns=top6_features)

rf_top = RandomForestClassifier(random_state=42)
rf_top.fit(xtrain_top, ytrain)
y_pred_top = rf_top.predict(xtest_top)

print('\n Top 10 Feature Model Evaluation:')
print(classification_report(ytest, y_pred_top))
print('Accuracy:', accuracy_score(ytest, y_pred_top))
print('F1 Score:', f1_score(ytest, y_pred_top, average='weighted'))
print('Precision:', precision_score(ytest, y_pred_top, average='weighted'))
print('Recall:', recall_score(ytest, y_pred_top, average='weighted'))
y_proba = rf_top.predict_proba(xtest_top)[:, 1]
print('ROC-AUC:', roc_auc_score(ytest, y_proba))
print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred_top))
```

WITH OVERSAMPLING

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, confusion_matrix,f1_score, precision_score, recall_score, roc_auc_score)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

X_clean = X.dropna()
y_clean = y.loc[X_clean.index]

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_clean, y_clean)
print('Resampled dataset shape:', Counter(y_res))

xtrain, xtest, ytrain, ytest = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

rf_temp = RandomForestClassifier(random_state=42)
rf_temp.fit(xtrain, ytrain)

importances = rf_temp.feature_importances_
forest_importances = pd.Series(importances, index=X_res.columns)
top7_features = forest_importances.sort_values(ascending=False).head(7).index.tolist()

print("\n Top 7 Important Features:")
print(top7_features)

std = np.std([tree.feature_importances_ for tree in rf_temp.estimators_], axis=0)
fig, ax = plt.subplots(figsize=(10, 6))
forest_importances[top7_features].plot.bar(yerr=std[:7], ax=ax)
ax.set_title("Top 7 Feature Importances")
ax.set_ylabel("Importance")
plt.tight_layout()
plt.show()

scaler = MinMaxScaler()
xtrain_top = pd.DataFrame(scaler.fit_transform(xtrain[top7_features]), columns=top7_features)
xtest_top = pd.DataFrame(scaler.transform(xtest[top7_features]), columns=top7_features)

rf_top = RandomForestClassifier(random_state=42)
rf_top.fit(xtrain_top, ytrain)
y_pred_top = rf_top.predict(xtest_top)

print('\n Top 10 Feature Model Evaluation:')
print(classification_report(ytest, y_pred_top))
print('Accuracy:', accuracy_score(ytest, y_pred_top))
print('F1 Score:', f1_score(ytest, y_pred_top, average='weighted'))
print('Precision:', precision_score(ytest, y_pred_top, average='weighted'))
print('Recall:', recall_score(ytest, y_pred_top, average='weighted'))
y_proba = rf_top.predict_proba(xtest_top)[:, 1]
print('ROC-AUC:', roc_auc_score(ytest, y_proba))
print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred_top))
```
