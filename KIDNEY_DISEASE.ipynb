FILLING MISSING VALUES - MODEL BASED IMPUTATION

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

df_model_impute = df_encoded.copy()
target_column = 'classification'
missing_cols = df_model_impute.columns[df_model_impute.isnull().any()]
missing_cols = [col for col in missing_cols if col != target_column]

# Loop through each column with missing values and perform model-based imputation
for col in missing_cols:
    print(f"\n🔄 Imputing column: {col}")

    # Separate rows with and without missing values for this column
    df_not_null = df_model_impute[df_model_impute[col].notnull()]
    df_null = df_model_impute[df_model_impute[col].isnull()]

    # Drop the target column from training features to avoid leakage
    X_train = df_not_null.drop(columns=[col, target_column])  # Exclude target column
    y_train = df_not_null[col]
    X_pred = df_null.drop(columns=[col, target_column])  # Exclude target column

    # Align the columns of X_pred with X_train to ensure matching feature set
    X_pred = X_pred.reindex(columns=X_train.columns, fill_value=0)

    # Check the number of unique values in the column to decide whether it's classification or regression
    num_unique_values = y_train.nunique()

    if num_unique_values == 2:  # Binary classification (exactly 2 unique values, e.g., 0 and 1)
        print(f"📊 Treating column {col} as a binary classification problem.")
        model = RandomForestClassifier(random_state=42)
        y_train = y_train.astype(int)  # Ensure target column is integer for classification
    elif num_unique_values <= 10:  # Multi-class classification (<= 10 unique values)
        print(f"📊 Treating column {col} as a multi-class classification problem.")
        model = RandomForestClassifier(random_state=42)
        y_train = y_train.astype(int)  # Ensure target column is integer for classification
    else:  # Regression (more than 10 unique values)
        print(f"📈 Treating column {col} as a regression problem.")
        model = RandomForestRegressor(random_state=42)
        y_train = y_train.astype(float)  # Ensure target column is float for regression

    # Train the model on non-null data
    model.fit(X_train, y_train)

    # Predict the missing values for the null rows
    y_pred = model.predict(X_pred)

    # If the column is categorical (binary or multi-class), round the predictions
    if num_unique_values <= 10:  # Round predictions for classification (binary or multi-class)
        y_pred = y_pred.round().astype(int)  # Round predictions to nearest integer for classification

    # Fill the missing values in the dataframe with the predicted values
    df_model_impute.loc[df_model_impute[col].isnull(), col] = y_pred

# Final check to ensure there are no missing values
print("\n✅ All missing values filled:", df_model_impute.isnull().sum().sum() == 0)
```

```python
import numpy as np

def cap_outliers_iqr(df, columns):
  for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    if lower_bound < 0:
        lower_bound = 0
    original_min = df[col].min()
    original_max = df[col].max()
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    print(f"{col}: capped below {lower_bound:.2f} and above {upper_bound:.2f} | Original Min: {original_min:.2f}, Max: {original_max:.2f}")
  return df
numeric_cols = df_model_impute.select_dtypes(include=['int64','float64']).columns.tolist()
numeric_but_categorical = [col for col in numeric_cols if df_model_impute[col].nunique() <= 10]
continuous_numeric_cols = [col for col in numeric_cols if col not in numeric_but_categorical]
df_model_impute = cap_outliers_iqr(df_model_impute, continuous_numeric_cols)
```

**MACHINE LEARNING MODELS**

ORIGINAL DATA (WITHOUT SAMPLING)

```python
X=df_encoded.drop("classification",axis=1)
y=df_encoded["classification"]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, accuracy_score, confusion_matrix,f1_score, precision_score, recall_score, roc_auc_score)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import joblib
import warnings
warnings.filterwarnings("ignore")

#Drop missing values
X_clean = X.dropna()
y_clean = y.loc[X_clean.index]

#Train-test split
xtrain, xtest, ytrain, ytest = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42, stratify=y_clean)

#Feature Importance
rf_temp = RandomForestClassifier(random_state=42)
rf_temp.fit(xtrain, ytrain)

#Top 10 features
importances = rf_temp.feature_importances_
forest_importances = pd.Series(importances, index=X_clean.columns)
top10_features = forest_importances.sort_values(ascending=False).head(10).index.tolist()

print("\nTop 10 important features:")
print(top10_features)

#Plot feature importances
std = np.std([tree.feature_importances_ for tree in rf_temp.estimators_], axis=0)
fig, ax = plt.subplots(figsize=(10, 6))
forest_importances[top10_features].plot.bar(yerr=std[:10], ax=ax)
ax.set_title("Top 10 Feature Importances")
ax.set_ylabel("Importance")
plt.tight_layout()
plt.show()

#Scaling Only Top 10 Features
scaler = MinMaxScaler()
xtrain_top = pd.DataFrame(scaler.fit_transform(xtrain[top10_features]), columns=top10_features)
xtest_top = pd.DataFrame(scaler.transform(xtest[top10_features]), columns=top10_features)

#Train Final Model
rf_top = RandomForestClassifier(random_state=42)
rf_top.fit(xtrain_top, ytrain)
y_pred_top = rf_top.predict(xtest_top)

#Evaluation
print('\nTop 10 Feature Model Evaluation:')
print(classification_report(ytest, y_pred_top))
print('Accuracy:', accuracy_score(ytest, y_pred_top))
print('F1 Score:', f1_score(ytest, y_pred_top, average='weighted'))
print('Precision:', precision_score(ytest, y_pred_top, average='weighted'))
print('Recall:', recall_score(ytest, y_pred_top, average='weighted'))
y_proba = rf_top.predict_proba(xtest_top)[:, 1]
print('ROC-AUC:', roc_auc_score(ytest, y_proba))
print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred_top))
```
